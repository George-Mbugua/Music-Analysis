Analyzing the distribution and relationships between variables such as key, loudness, mode, and tempo to identify patterns and trends in the musical attributes of songs.

1. Key: refers to the tonal center or tonic note around which a musical composition is organized.

2. Loudness: is a measure of the perceived volume of a song.

3. Mode: refers to the type of scale from which the composition’s melody and harmonies are derived.

4. Tempo: is the speed at which a piece of music is played, typically measured in beats per minute (BPM).

Calculating mean, median, mode, variance, standard deviation for key, loudness, mode, and tempo.


```python

def calculate_statistics(data):
    # Calculate the mean for each musical characteristic
    mean_key = data['Key'].mean()
    mean_loudness = data['Loudness'].mean()
    mean_mode = data['Mode'].mean()
    mean_tempo = data['Tempo'].mean()

    # Calculate the median for each musical characteristic
    median_key = data['Key'].median()
    median_loudness = data['Loudness'].median()
    median_mode = data['Mode'].median()
    median_tempo = data['Tempo'].median()

    # Calculate the mode for each musical characteristic
    mode_key = data['Key'].mode()[0]
    mode_loudness = data['Loudness'].mode()[0]
    mode_mode = data['Mode'].mode()[0]
    mode_tempo = data['Tempo'].mode()[0]

    # Calculate the variance for each musical characteristic
    variance_key = data['Key'].var()
    variance_loudness = data['Loudness'].var()
    variance_mode = data['Mode'].var()
    variance_tempo = data['Tempo'].var()

    # Calculate the standard deviation for each musical characteristic
    std_dev_key = data['Key'].std()
    std_dev_loudness = data['Loudness'].std()
    std_dev_mode = data['Mode'].std()
    std_dev_tempo = data['Tempo'].std()

    return {
        'mean': {
            'Key': mean_key,
            'Loudness': mean_loudness,
            'Mode': mean_mode,
            'Tempo': mean_tempo,
        },
        'median': {
            'Key': median_key,
            'Loudness': median_loudness,
            'Mode': median_mode,
            'Tempo': median_tempo,
        },
        'mode': {
            'Key': mode_key,
            'Loudness': mode_loudness,
            'Mode': mode_mode,
            'Tempo': mode_tempo,
        },
        'variance': {
            'Key': variance_key,
            'Loudness': variance_loudness,
            'Mode': variance_mode,
            'Tempo': variance_tempo,
        },
        'standard_deviation': {
            'Key': std_dev_key,
            'Loudness': std_dev_loudness,
            'Mode': std_dev_mode,
            'Tempo': std_dev_tempo,
        }
    }

statistics = calculate_statistics(music)

for key in statistics:
  print(key)
  for sub_key in statistics[key]:
      print(f'  {sub_key}: {statistics[key][sub_key]}')
```

From the results above there are a few observations:

1.Key: The mean and median values are close, indicating that the data is relatively symmetrically distributed. The mode is 0, indicating that the most common key in the dataset is 0. The standard deviation is relatively large compared to the range of possible key values (0-11), indicating that there is a wide spread of key values in the dataset.

2.Loudness: The mean and median values are relatively close, but the mean is lower, indicating that the data may be slightly skewed to the left. The standard deviation is relatively large, indicating that there is a wide spread of loudness values in the dataset.

3.Mode: The mean value is close to 1, indicating that a majority of songs in the dataset are in a major key. The mode is 1, indicating that the most common mode in the dataset is major. The standard deviation is relatively small, indicating that there is not much variation in mode values in the dataset.

4.Tempo: The mean and median values are close, indicating that the data is relatively symmetrically distributed. The mode is close to 120, indicating that the most common tempo in the dataset is around 120 BPM. The standard deviation is relatively large, indicating that there is a wide spread of tempo values in the dataset.

#### 2.Identifying patterns in user listening behavior by analyzing variables such as msPlayed and genre.

a.Analyzing  Genre vs Msplayed

```python
# Calculate the total time listened for each track
total_time_by_track = music.groupby('Trackname')['Msplayed'].sum()

# The top 10 most listened tracks
top_tracks = total_time_by_track.sort_values(ascending=False).head(10)
top_tracks_df = top_tracks.to_frame().reset_index()
top_tracks_df.index += 1
print(top_tracks_df)
#Plot the top 10 most listened tracks
plt.figure(figsize=(10, 6))
sns.barplot(x=top_tracks.values, y=top_tracks.index, orient='h')
plt.title('Top 10 Most Listened Tracks')
plt.xlabel('Total Time Listened (ms)')
plt.ylabel('Track Name')
plt.show()

# Calculate the total time listened for each genre
total_time_by_genre = music.groupby('Genre')['Msplayed'].sum()
total_time_by_genre = total_time_by_genre.sort_values(ascending=False).head(15)
total_time_by_genre_df = total_time_by_genre.to_frame().reset_index()
total_time_by_genre_df.index += 1
print(total_time_by_genre_df)

# Plot the total time listened for each genre
plt.figure(figsize=(10, 6))
sns.barplot(x=total_time_by_genre.index, y=total_time_by_genre.values)
plt.title('Total Time Listened by Genre')
plt.xlabel('Genre')
plt.xticks(rotation=90)
plt.ylabel('Total Time Listened (ms)')
plt.show()
```

Analyzing Genre and Danceability


```python
# Calculate the average danceability for each genre
avg_danceability_by_genre = music.groupby('Genre')['Danceability','Msplayed'].mean()

# Select the top 15 most danceable genres
top_15_avg_danceability_by_genre = avg_danceability_by_genre.sort_values(by ='Danceability', ascending=False).head(20)
# Create a DataFrame with the results
results = pd.DataFrame(top_15_avg_danceability_by_genre)

# Display the results in a DataFrame
print(results)
```

```python
# Create a bar chart of the average danceability by genre
results.plot(kind='barh', y='Danceability', title='Average Danceability by Genre', figsize=(10, 6))

# Add labels to the x and y axes
plt.xlabel('Average Danceability')
plt.ylabel('Genre')
# Show the plot
plt.show()
```

According to the bar graph, Canadian old school hip hop, Trap queen and Russian drill have the most danceability while estonian electonic and Turkish edm have the least danceability.


c. Analyzing Genre vs energy,key,loudness,mode,speechiness,liveness,valence,tempo

```python
# Calculate the total time listened for each genre
total_time_by_genre = music.groupby('Genre')['Msplayed'].sum()

# Select the top 20 genres based on total time listened
top_genres = total_time_by_genre.sort_values(ascending=False).head(15).index

# Filter the data to only include the top 15 genres
filtered_data = music[music['Genre'].isin(top_genres)]

# Calculate the average value for each audio feature by genre
results = filtered_data.groupby('Genre').agg({'Energy': 'mean', 'Key': 'mean', 'Loudness': 'mean', 'Mode': 'mean', 'Speechiness': 'mean', 'Liveness': 'mean', 'Valence': 'mean', 'Tempo': 'mean'})

# Display the results in a DataFrame
print(results)
```

```python
# Create a bar plot to compare the average energy by genre for the top 15 genres
plt.figure(figsize=(10, 6))
sns.barplot(data=filtered_data, x='Genre', y='Energy')
plt.title('Average Energy by Genre for Top 15 Genres')
plt.xlabel('Genre')
plt.ylabel('Average Energy')
plt.xticks(rotation=90)
plt.show()

# Create a box plot to compare the distribution of loudness by genre for the top 15 genres
plt.figure(figsize=(10, 6))
sns.boxplot(data=filtered_data, x='Genre', y='Loudness')
plt.title('Loudness by Genre for Top 15 Genres')
plt.xlabel('Genre')
plt.ylabel('Loudness')
plt.xticks(rotation=90)
plt.show()

# Create a violin plot to compare the distribution of tempo by genre for the top 15 genres
plt.figure(figsize=(10, 6))
sns.violinplot(data=filtered_data, x='Genre', y='Tempo')
plt.title('Tempo by Genre for Top 15 Genres')
plt.xlabel('Genre')
plt.ylabel('Tempo')
plt.xticks(rotation=90)
plt.show()
```

From the graphs above we can deduce that for the top 15 tracks:

Energy: The genre with the highest average energy is ‘edm’ with a value of 0.752600, while the genre with the lowest average energy is ‘german soundtrack’ with a value of 0.111989.

Key: The genre with the highest average key is ‘japanese teen pop’ with a value of 6.600000, while the genre with the lowest average key is ‘baroque’ with a value of 4.250000.

Loudness: The genre with the highest average loudness is ‘edm’ with a value of -4.847980, while the genre with the lowest average loudness is ‘german soundtrack’ with a value of -24.643206.

Mode: The genre with the highest average mode is ‘baroque’ with a value of 0.875000, while the genre with the lowest average mode is ‘japanese teen pop’ with a value of 0.200000.

Speechiness: The genre with the highest average speechiness is ‘la pop’ with a value of 0.096652, while the genre with the lowest average speechiness is ‘japanese teen pop’ with a value of 0.033840.
