```python

def calculate_statistics(data):
    # Calculate the mean for each musical characteristic
    mean_key = data['Key'].mean()
    mean_loudness = data['Loudness'].mean()
    mean_mode = data['Mode'].mean()
    mean_tempo = data['Tempo'].mean()

    # Calculate the median for each musical characteristic
    median_key = data['Key'].median()
    median_loudness = data['Loudness'].median()
    median_mode = data['Mode'].median()
    median_tempo = data['Tempo'].median()

    # Calculate the mode for each musical characteristic
    mode_key = data['Key'].mode()[0]
    mode_loudness = data['Loudness'].mode()[0]
    mode_mode = data['Mode'].mode()[0]
    mode_tempo = data['Tempo'].mode()[0]

    # Calculate the variance for each musical characteristic
    variance_key = data['Key'].var()
    variance_loudness = data['Loudness'].var()
    variance_mode = data['Mode'].var()
    variance_tempo = data['Tempo'].var()

    # Calculate the standard deviation for each musical characteristic
    std_dev_key = data['Key'].std()
    std_dev_loudness = data['Loudness'].std()
    std_dev_mode = data['Mode'].std()
    std_dev_tempo = data['Tempo'].std()

    return {
        'mean': {
            'Key': mean_key,
            'Loudness': mean_loudness,
            'Mode': mean_mode,
            'Tempo': mean_tempo,
        },
        'median': {
            'Key': median_key,
            'Loudness': median_loudness,
            'Mode': median_mode,
            'Tempo': median_tempo,
        },
        'mode': {
            'Key': mode_key,
            'Loudness': mode_loudness,
            'Mode': mode_mode,
            'Tempo': mode_tempo,
        },
        'variance': {
            'Key': variance_key,
            'Loudness': variance_loudness,
            'Mode': variance_mode,
            'Tempo': variance_tempo,
        },
        'standard_deviation': {
            'Key': std_dev_key,
            'Loudness': std_dev_loudness,
            'Mode': std_dev_mode,
            'Tempo': std_dev_tempo,
        }
    }

statistics = calculate_statistics(music)

for key in statistics:
  print(key)
  for sub_key in statistics[key]:
      print(f'  {sub_key}: {statistics[key][sub_key]}')
```

```python
# Bar Charts
plt.bar(statistics['mean'].keys(), statistics['mean'].values())
plt.title('Means')
plt.show()


plt.bar(statistics['median'].keys(), statistics['median'].values())
plt.title('Medians')
plt.show()


plt.bar(statistics['mode'].keys(), statistics['mode'].values())
plt.title('Modes')
plt.show()


plt.bar(statistics['variance'].keys(), statistics['variance'].values())
plt.title('Variances')
plt.show()
```
