## 4. Outliers

As seen the average duration of a song in the dataset is 202,931 milliseconds(3 minutes and 23 seconds), with a maximum of 76 minutes and the lowest 10 seconds. I will filter the dataframe to show only those that are 5 minutes and less than 60 seconds.

```python
#Plotting boxplot to visualize the outliers
sns.boxplot(data = music, x = 'Msplayed', showfliers=True);
```

```python
#Defining the duration thresholds in milliseconds
min_duration = 60 * 1000  # 60 seconds
max_duration = 5 * 60 * 1000  # 5 minutes
#Finding rows where the msPlayed value is within the required range
music.loc[(music['Msplayed'] >= min_duration) & (music['Msplayed'] <= max_duration)]
```

```python
#Dropping outliers
music = music.loc[(music['Msplayed'] >= min_duration) & (music['Msplayed'] <= max_duration)]
music.shape
```
