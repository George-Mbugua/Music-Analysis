### 4. Outliers

As seen the average duration of a song in the dataset is 202,931 milliseconds(3 minutes and 23 seconds), with a maximum of 76 minutes and the lowest 10 seconds. I will filter the dataframe to show only those that are 5 minutes and less than 60 seconds.

```python
#Plotting boxplot to visualize the outliers
sns.boxplot(data = music, x = 'Msplayed', showfliers=True);
```

```python
#Defining the duration thresholds in milliseconds
min_duration = 60 * 1000  # 60 seconds
max_duration = 5 * 60 * 1000  # 5 minutes
#Finding rows where the msPlayed value is within the required range
music.loc[(music['Msplayed'] >= min_duration) & (music['Msplayed'] <= max_duration)]
```

```python
#Dropping outliers
music = music.loc[(music['Msplayed'] >= min_duration) & (music['Msplayed'] <= max_duration)]
music.shape
```

```python
#Plotting boxplot to visualize the outliers again
sns.boxplot(data = music, x = 'Msplayed', showfliers=True);
```

### 1. What is the most popular genre of music?

Most popular genres in terms of Msplayed

```python
#Getting the top 10 genres by appearance on the dataframe
popular_genres = music['Genre'].value_counts().head(10)
print('The top 10 genres are:')
print(popular_genres)
```

Most popular genres in terms of Msplayed

```python
#Grouping the data by Genre and Msplayed column within each group
genre_msplayed = music.groupby('Genre')['Msplayed'].sum()
```

```python
#Sorting the resulting Series in descending order
genre_msplayed = genre_msplayed.sort_values(ascending=True).head(10)
genre_msplayed
```

```python
#Creating a bar chart to visualize the most popular genres in terms of Msplayed
fig = plt.figure()
genre_msplayed.plot( x='Msplayed', y='Genre', kind='barh', )
plt.title('Most Popular Genres in Terms of Msplayed')
plt.xlabel('Total Msplayed')
plt.ylabel('Genre')
plt.show()
fig.savefig('Images/Most Popular Genres in Terms of Msplayed')
```

Observations:
- The most popular genre is pop, with a total of 2,463,1007, followed by filmi and the third most popular genre is alt z.
- There appears to be different kinds of pop eg, art pop, baroque pop

### 2. What are the characteristics of popular songs?

First I will create a correlation matrix to better understand the data


```python
#Creating the correlation matrix
corr= music.corr()
corr
```

```python
#Visualizing~ the correlation matrix using a heatmap
plt.figure(figsize=(20, 20))
sns.heatmap(corr, annot=True)
plt.show()
```

Observations:
1. Energy and Loudness have a positive correlation with Msplayed. This suggests that songs with higher energy and loudness levels tend to be more popular.

2. Acousticness and Instrumentalness have a negative correlation with Msplayed This suggests that songs with higher levels of acousticness and instrumentalness tend to be less popular.

3. Danceability, Valence, and Tempo have relatively weak correlations with Msplayed, with correlation coefficients of 0.011106, 0.067295, and -0.003865, respectively. This suggests that these characteristics may not be as strongly associated with the popularity of songs.

4. Duration_ms has the strongest positive correlation with Msplayed, with a correlation coefficient of 0.399180. This suggests that longer songs tend to be more popular.
